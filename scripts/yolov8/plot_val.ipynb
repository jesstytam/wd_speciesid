```python
from PIL import Image, ImageDraw, ImageFont

predictions_path = '/home/jess2/wd_speciesid/scripts/yolov8/runs/detect/val2/predictions.json'

for image in predictions_path:

    # Load the image
    image_path = '/home/jess2/data/wild_deserts/Beyond the Fence- Tagged/images/PCAM01/2021-07-24 20-48-10 M 3_3.JPG' #CHANGE
    image = Image.open(image_path)

    # Create a drawing object
    draw = ImageDraw.Draw(image)

    # Define the bounding box parameters (x, y, width, height)
    x, y, width, height = image['bbox']  # Replace with your bounding box coordinates

    # Define the outline color (red in this example)
    outline_color = (255, 0, 0)  # RGB color

    # Draw a rectangle on the image
    draw.rectangle([x, y, x + width, y + height], outline=outline_color, width=3)

    # Define the label text and text color
    label_text = "'category_id': 4"  # Replace with your label text
    text_color = (255, 255, 255)  # RGB color

    # Load a font (you may need to specify the path to your font file)
    font_size = 30
    font = ImageFont.load_default()
    # font = ImageFont.truetype("arial.ttf", size=20)

    # Calculate text position
    text_x = x
    text_y = y - 30  # Adjust this value to set the text above the bounding box

    # Draw the label text on the image
    draw.text((text_x, text_y), label_text, fill=text_color, font=font)

    # Save or display the image
    # image.save('/path/to/output_image.jpg')
    image.show()
```
