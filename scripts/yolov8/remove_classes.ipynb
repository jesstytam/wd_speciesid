```python
old_to_new_mapping = {
    11: 1,
    12: 5,
    14: 6
}

classes_to_remove = [1, 5, 6, 13]
```

```python
#remove classes
for item in train['annotations']:
    if item['category_id'] in classes_to_remove:
        train['annotations'].remove(item)
    else:
        pass
```

```python
#replace classes
for item in train['annotations']:
    if item['category_id'] in old_to_new_mapping.keys():
        item['category_id'] = old_to_new_mapping.get(item['category_id'])
```

```python
ids = []
for item in train['annotations']:
    ids.append(item['category_id'])
set(ids)
```

```python
image_id = []
for image in train['annotations']:
    image_id.append(image['image_id'])
print(len(image_id))
```

```python
len(train['images'])
```

```python
for image in train['images']:
    if image['id'] not in image_id:
        train['images'].remove(image)
```

```python
for cat in train['categories']:
    if cat['id'] in classes_to_remove:
        train['categories'].remove(cat)
train['categories']
```

```python
#replace classes
for cat in train['categories']:
    if cat['id'] in old_to_new_mapping.keys():
        cat['id'] = old_to_new_mapping.get(cat['id'])
train['categories']
```
