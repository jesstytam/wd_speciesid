# 2. Clean

## 2.1 update categories

```python
#get classes
import pandas as pd
labels = pd.read_excel('~/wd_speciesid/data/raw/BtF-information/Beyond_the_Fence_ALL_TAGS_to_Feb22.xlsx', header=1)
```

```python
spp = labels['Species'].unique()
print(spp)
```

```python
# ids = {}
# names = {}
new_list = []

i=0
for sp in spp:
    i = i+1
    new_dict = {'id':i, 'name':sp}
    new_list.append(new_dict)

print(new_list)
```

```python
dirty_coco['categories'][1:]
```

```python
dirty_coco['categories'][1:] = new_list
dirty_coco['categories']
```

## 2.2 update annotation ids

```python
dirty_coco['annotations'][0]
```

```python
labels['Species']
```

```python
for i in range(len(labels)):
    
    #get image_id
    img_name = labels['Image Name'][i][:-4].replace(' ', '_')
    site = labels['Site'][i]
    id_from_xlsx = site+'_'+img_name

    for annotation in dirty_coco['annotations']:

        #match image id to to img name
        if annotation['image_id'] == id_from_xlsx:
            new_cat_name = labels['Species'][i]

            # Create a dictionary to map name to id
            name_to_id = {category['name']: category['id'] for category in dirty_coco['categories']}

            #get new id
            matching_id = name_to_id.get(new_cat_name)
            if matching_id is not None:
                annotation['category_id'] = matching_id
```

# 3. Checking!

```python
clean_coco = json.load(open('/home/jess2/wd_speciesid/data/intermediate/coco.json', 'r'))
clean_coco
```

```python
clean_coco['categories']
```

```python
# Create a dictionary to store the counts
id_counts = {}

# Loop through the annotations and count occurrences of each ID
for annotation in clean_coco['annotations']:
    category_id = annotation['category_id']
    if category_id in id_counts:
        id_counts[category_id] += 1
    else:
        id_counts[category_id] = 1

# Print the counts
for category_id, count in id_counts.items():
    category_name = name_to_id.get(category_id)  # Get category name from the dictionary
    print(f"Category '{category_name}' (ID {category_id}): {count} occurrences")
```
